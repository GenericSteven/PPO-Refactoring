```python
import pandas as pd
import matplotlib.pyplot as plt

# File path (update this path to where your CSV file is located)
file_path = './logs/levelbeats.csv'

# Read the newline-separated CSV file into a pandas DataFrame
# 68 timesteps per second
data = pd.read_csv(file_path, header=None, names=['Win Times'])

# Pass 1 Fastest time: 55.65 seconds
# Pass 2 Fastest time: ??
# Plot the data
lowest_value = data['Win Times'].min()
print(f'Fastest time: {round(lowest_value / 68,2)} seconds')
plt.plot(data['Win Times'] / 68)

# Set the title and y-axis label
plt.title('Win Times Over 1,000,000 Timesteps')
plt.ylabel('Timesteps needed to win')
plt.xlabel('Number of wins')

# Display the plot
plt.show()
```
